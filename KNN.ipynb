# K Nearest Neighbors 


```python
#import all the libraries we need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#This is show us a edges in the graph
plt.rcParams["patch.force_edgecolor"] = True
```

```python
#Load the data
datas = pd.read_csv("KNN_project_Data")
#Check the head of the dataframe.
datas.head()
```

```python
#check if there is any empty data
sns.heatmap(datas.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
#use pairplot for data visualization
sns.pairplot(datas,hue="TARGET CLASS")
sns.plt.show()
```

```python
#Standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
#Create a StandardScaler() object called scaler.
scaler = StandardScaler()
```

```python
scaler.fit(datas.drop("TARGET CLASS",axis=1))
```

```python
scaled_feature = scaler.transform(datas.drop("TARGET CLASS", axis=1))
#scaled_feature
```

```python
datas.columns[:-1]
```

```python
#Convert the scaled features to a dataframe 
data = pd.DataFrame(scaled_feature,columns=datas.columns[:-1])
data.head()
```

# Train Test Split


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_feature,datas["TARGET CLASS"],
                                                test_size=0.30)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    predic_i = knn.predict(X_test)
    error_rate.append(np.mean(predic_i != y_test))
```

```python
#Plot the graph of error rate to see the optimal k value
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color="blue",linestyle="dashed",marker="o",
        markerfacecolor="red",markersize=10)
plt.title("Error Rate vs. K Value")
plt.xlabel("K Value")
plt.ylabel("Error Rate")
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#since best k value is 25
knn = KNeighborsClassifier(n_neighbors=32)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=32')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
